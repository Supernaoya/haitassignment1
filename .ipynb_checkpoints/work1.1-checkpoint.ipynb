```python
#カテゴリデータの前処理：ダミー変数化
df_train = pd.get_dummies(df_train)
```

```python
pd.get_dummies(df_train)#ダミー変数化したあとに表示
```

```python
# 説明変数と目的変数を指定
# trainデータを用意
# DataFrameの.valuesメソッドでarrayに変換
X = df_train.loc[:, ['TotalBsmtSF', 'GrLivArea']].values
```

```python
X
```

```python
# testデータを用意
# DataFrameの.valuesメソッドでarrayに変換
y = df_train.loc[:, ['SalePrice']].values
```

```python
y
```

```python
# ホールド・アウト法によるデータの分割(DLしたtestにSaleprice入ってないから、)
# (X, y, テストにどの割合を割くか, 分割のしかたを決める乱数生成器)
# テストデータの割合は、test_size引数に指定することで変更可能です（今回は、train:test = 7:3）
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# データ全体のサンプル数を確認
print(X.shape)
print(y.shape)
```

```python
# trainデータのサンプル数を確認
print(X_train.shape)
print(y_train.shape)
```

```python
# testデータのサンプル数を確認
print(X_test.shape)
print(y_test.shape)
```

```python
# 重回帰の線形回帰を実行
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# 重回帰線形モデルの学習
# trainデータのみを用いていることに注意
lr.fit(X_train, y_train)
```

```python
# 定数項を出力
lr.intercept_
```

```python
# 偏回帰係数を出力
# ['TotalBsmtSF'の係数, 'GrLivArea'の係数]
lr.coef_
```

```python
# y=5.4219492+0.084953x1+0.82853723x2
```

```python
# 決定係数をtrainとtestに分けて出力
# 過学習しているかどうかを確認
print('R^2')
print('train: %.3f' % lr.score(X_train, y_train))
print('test : %.3f' % lr.score(X_test, y_test))
#学習させたモデルはあまり当てはまりが良いとは言えなさそう
```

```python
# 自由度調整済み決定係数
# (決定係数, trainまたはtestのサンプル数, 利用した特徴量の数)
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
```

```python
# 自由度調整済み決定係数を出力
print('adjusted R^2')
print('train: %3f' % adjusted(lr.score(X_train, y_train), len(y_train), 2))
print('test : %3f' % adjusted(lr.score(X_test, y_test), len(y_test), 2))
#結局、学習させたモデルはあまり当てはまりが良いとは言えなさそう
```
